## Plotting from pandas DataFrame

```python
car_sales = pd.read_csv("car-sales.csv.csv")
car_sales
```

```python
ts = pd.Series(np.random.randn(1000),
              index = pd.date_range("1/1/2020", periods = 1000))
ts = ts.cumsum()
ts.plot();
```

```python
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\,]', '')
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2020", periods = len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].astype(float).cumsum()
car_sales
```

```python
# For pandas DataFrame we can directly call .plot 
car_sales.plot(x = "Sale Date", y = "Total Sales");
```

```python
car_sales.plot(x = "Sale Date", kind="bar")
```

```python
car_sales["Odometer (KM)"].plot(kind = "hist")
```

```python
 car_sales["Odometer (KM)"].plot.hist(bins = 20)
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
heart_disease["age"].plot.hist(bins=50)
```

```python
heart_disease.plot.hist(figsize = (10,30));
```

```python
heart_disease.plot.hist(figsize = (10,30), subplots = True);
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
```

```python
over_50.head()
```

```python
# PyPlot Method

over_50.plot(kind = "scatter",
            x="age",
            y="chol",
            c="target")
```

```python
# OO method -- This method is preferred when plotting advanced graphs, here we can modify the axis and graph accodingly
fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = "scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
# ax.set_xlim([45,100])
```

```python
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
ax.set(title = "Heart Disease",
      xlabel = "Age",
      ylabel = "Cholestrol")
# Add a legend
ax.legend(*scatter.legend_elements(), title = "Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle = "--");
```

```python
over_50.head()
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols = 1,
                              figsize = (10,10))
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
ax0.set(title = "Heart Disease",
      xlabel = "Age",
      ylabel = "Cholestrol")

ax0.legend(*scatter.legend_elements(), title = "Target")
ax0.axhline(y = over_50["chol"].mean(),
           linestyle = "--")


scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])
ax1.set(title = "Heart Disease",
      xlabel = "Age",
      ylabel = "Max Heart Rate")

ax1.legend(*scatter.legend_elements(), title = "Target")
ax1.axhline(y = over_50["thalach"].mean(),
           linestyle = "--");

# Add title to the whole figure

fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```
